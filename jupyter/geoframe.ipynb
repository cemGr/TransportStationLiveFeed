```python
import geopandas as gpd
from shapely.geometry import Point

# Load datatripoints csv file
datatripoints = gpd.read_file('cleaned_trip_data.csv')
```

```python
datatripoints
```

```python
# GeoDataFrame mit Start- und Endpunkt
gdf = gpd.GeoDataFrame(datatripoints, geometry=gpd.points_from_xy(datatripoints["start_lon"], datatripoints["start_lat"]), crs="EPSG:4326")
gdf_end = gpd.GeoDataFrame(datatripoints, geometry=gpd.points_from_xy(datatripoints["end_lon"], datatripoints["end_lat"]), crs="EPSG:4326")

# In ein metrisches CRS umprojizieren (hier EPSG:3857)
gdf_m = gdf.to_crs(epsg=3857)
gdf_end_m = gdf_end.to_crs(epsg=3857)

# Distanz in Metern
dist_m = gdf_m.geometry.distance(gdf_end_m.geometry)

print(dist_m[0], "Meter")
dist_m
```
